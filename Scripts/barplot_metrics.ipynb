```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "County": ["Tehama", "Lassen"],
    "MAE": [15.454531, 5.36],
    "RMSE": [19.413636, 5.36],
    "R2": [-0.275707, None]
}
df = pd.DataFrame(data)

# Plotting
metrics = ["MAE", "RMSE", "R2"]
for metric in metrics:
    plt.figure(figsize=(8, 6))
    plt.bar(df["County"], df[metric], alpha=0.7, label=metric)
    plt.title(f"{metric} by County")
    plt.xlabel("County")
    plt.ylabel(metric)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

colors = ["skyblue", "lightgreen", "salmon", "violet"]

for i, metric in enumerate(metrics):
    plt.figure(figsize=(8, 6))
    plt.bar(df["County"], df[metric], color=colors[i], alpha=0.7, label=metric)
    plt.title(f"{metric} by County")
    plt.xlabel("County")
    plt.ylabel(metric)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
